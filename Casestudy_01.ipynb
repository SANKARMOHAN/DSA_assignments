```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1) Read the dataset to the python environment.

```python
data1 = pd.read_csv('cars_data.csv')
data1
```

A dataset having 10000 rows and 20 columns are stored.

# 2) Check for the null values present in the dataset.

```python
data1.info()
```

No null values are present in the dataset

# 3) Plot a bar graph of male vs female buyers participated in the sales.

```python
plt.figure(figsize=(10,8))
sns.countplot(x='Buyer Gender',data=data1,palette=["#FF6F91", "#2C73D2"])
plt.title('Female vs male',fontsize=25,fontweight='bold')
plt.grid(True)
```

From the graph, the total no. of female participants are more than that of male participants.

# 4) Find the top 5 cars based on their sales price.

```python
data1.sort_values('Sale Price',axis=0,ascending=False).head()
```

The top 5 cars are Volkswagen Jetta III, Audi A6, Lexus LX, Mazda MX-6 and Lexus ES.
