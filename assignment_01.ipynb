1. Read the dataset to python environment.

```python
#importing numpy,pandas, matplotlib & seaborn libraies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#dataset reading
iris = pd.read_excel('iris.xls')
iris
```

5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
#Histogram
plt.figure()
plt.hist(iris['SL'],color='blue',rwidth=0.5,density='true')
plt.title('Visualization of SL',fontsize=16)
plt.xticks()
plt.yticks()
```

```python
#Scatter Plot
plt.figure(figsize=(10,12))
plt.subplot(2,1,1)
plt.scatter(iris['SL'],iris['SW'],s=20,c='red')
plt.title('Plot of SL vs SW', fontsize=20)
plt.xlabel('SL',fontsize=16)
plt.ylabel('SW',fontsize=16,rotation=0)
plt.subplot(2,1,2)
plt.scatter(iris['PL'],iris['PW'],s=20,c='violet')
plt.title('Plot of PL vs PW', fontsize=20)
plt.xlabel('PL',fontsize=16)
plt.ylabel('PW',fontsize=16,rotation=0)
plt.tight_layout()
```

```python
#Pair Plot
sns.pairplot(iris,hue='Classification')
```

```python
#Heatmap Plot
sns.heatmap(iris.corr(),annot=True)
plt.xticks(rotation=45)
```

```python
#Box Plot
plt.figure(figsize=(15,15))
sns.violinplot(x='SL',y='Classification',data=iris)
```
